```python


i = lnks[0][1]

print("i",i)
#get the profile URL
driver.get(i)
time.sleep(random.uniform(2.5,4.9))

#assigning the source code for the web page to variable set
sel = Selector(text=driver.page_source)
print("sel",sel)
#xpath to extract the text from the class containing the name
name = driver.find_element(By.XPATH,'//h1[contains(@class, "text-heading-xlarge inline t-24 v-align-middle break-words")]').text
#name = sel.xpath('//*[starts-with(@class,"text-heading-xlarge inline t-24 v-align-middle break-words")]/text()').extract_first()
print("name", name)

#if name exists
if name:
    # .strip() will remove the new line /n and white spaces
    name = name.strip()



# Use driver.find_element_by_xpath to locate the element
job_title_element = driver.find_element(By.XPATH,'//div[contains(@class, "text-body-medium break-words")]').text
print("Job job_title_element:", job_title_element)
# Extract the text
job_title = job_title_element.strip() if job_title_element else None

# Print the result
print("Job Title:", job_title)

name = validate_field(name)
job_title = validate_field(job_title)
linkedin_url = i


#printing the output
print("\n")
print("Name: " + name)
print("Job Title: " + job_title)
print("URL: " + linkedin_url)
print('\n')

data = {
    'Name':name,
    'Job Title' : job_title,
    'URL' : linkedin_url
}
Jobdata.append(data)


        
        
        
        
        
```
